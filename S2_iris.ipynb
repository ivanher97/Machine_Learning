```python
# Nombres de las columnas
col_name = ['sepal lenght', 'sepal width', 'petal length', 'petal width', 'class']

# Cargamos la nueva fila encima de la informacion
df.columns = col_name

df.head()
```

```python
# Descargamos automaticamente un dataset predefinido desde un repositorio publico mantenido por el creador de seaborn. Mismos datos que el .csv
iris = sns.load_dataset("iris")
iris.head()
```

```python
# Generamos unas estadisticas descriptivas de las clumnas del dataFrame
# (count) numero de valores no nulos, (mean) media aritmetica, (std) desviacion standar , (min) valor minimo,
# (25%) indica que el 25% de los valores son menores o iguales que el indicado, (50%) indica el valor central,
# (75%) indica que el 75% de los valores son menores o iguales que el indicado, (max) valor maximo
df.describe()
```

```python
# Existen variaciones de datos debido a que los del df salen del archivo descargados y los de iris salen del repositorio
iris.describe()
```

```python
# Mostramos un resumen rapido del dataframe
print(iris.info())
```

```python
# Dividimos los datos por especies y indicamos que nos devuelva la cantidad de cada especie
print(iris.groupby('species').size())
```

```python
# Generacion de una matriz de graficos de dispersion
sns.pairplot(iris, hue='species', size=3, aspect=1)
```

```python
# Genera un histograma automático por cada columna numérica
iris.hist(edgecolor='black', linewidth=1.2, figsize=(12,8))
plt.show()
```

```python
# Generacion de graficos de violin 
# Indicamos las dimensiones de la imagen 12x8
plt.figure(figsize=(12,8))

# Indicamos el numero de filas, numero de columnas y el numero del grafico
plt.subplot(2,2,1)
# Dibuja un grafico de violin en el cual marcamos cuales son los ejes x e y
sns.violinplot(x = 'species', y = 'sepal_length', data = iris)

plt.subplot(2,2,2)
sns.violinplot(x = 'species', y = 'sepal_width', data = iris)

plt.subplot(2,2,3)
sns.violinplot(x = 'species', y = 'petal_length', data = iris)

plt.subplot(2,2,4)
sns.violinplot(x = 'species', y = 'petal_width', data = iris)
```

```python
# Mostramos un conjunto de boxplots indicando por que atributo se tienen que agrupar, ademas de indicar la dimension de la imagen
iris.boxplot(by='species', figsize=(12,8))
```

```python
# Generamos una matriz de diagramas de dispersion.
pd.plotting.scatter_matrix(iris, figsize=(12,10))
plt.show()
```
