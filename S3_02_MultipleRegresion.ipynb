## REGRESION MULTIPLE

- Técnica estadística utilizada para mopdelar la realción entre una variable dependiente (respuesta) y dos o más variables independientes (predictoras)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import sklearn
```

```python
# Cargamos el archivo como un DataFrame
df = pd.read_csv("/media/sf_MV_COMP/data/housing.data", delim_whitespace = True, header = None)
```

```python
# Establecemos los nombres de las columnas
col_name = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']
df.columns = col_name
df.head()
```

```python
df.shape
```

```python
x = df.iloc[:, :-1]
y = df['MEDV']
y
```

## STATSMODELS

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf

# Agregamos una columna con solo 1s a los datos, para asegurar que la formula tenga un termino fijo
x_constant = sm.add_constant(x)

# Convertimos los datos con la nueva columna en un formato DataFrame
pd.DataFrame(x_constant)
```

```python
# Linea para que nos muestre la documentacion
help(sm.OLS)
```

```python
# Crea un modelo que intente predecir los valores de 'y' usando los datos en x_constant
model = sm.OLS(y, x_constant)

# Usa los datos que le dimos para aprender cuál es la mejor combinación de factores que explican y
lr = model.fit()

# Muestra una tabla con:
    # Coeficientes → qué tan importante es cada factor (ej. habitaciones, edad del edificio, etc.)
    # Valores p → indican si cada factor es realmente útil o es solo ruido.
    # R² → qué tan bien se ajusta el modelo a los datos (si es cercano a 1, está muy bien).
    # Otros datos estadisticos
lr.summary()
```

## Resultados de la Regresión Lineal (`lr.summary()`)

El método `summary()` en `statsmodels` muestra una tabla con resultados estadísticos del modelo de regresión. Entre los más importantes:

### 📊 Tabla de Coeficientes
- **Coeficientes (`coef`)** → indican cuánto cambia la variable dependiente cuando el factor aumenta en una unidad, manteniendo los demás constantes.
- **Valores p (`P>|t|`)** → muestran si el coeficiente es estadísticamente significativo (usualmente si `p < 0.05`).
- **Intervalos de confianza (`[0.025, 0.975]`)** → rango dentro del cual probablemente se encuentre el valor real del coeficiente.

### 📈 Métricas del Modelo
- **R² (R-squared)** → mide qué tan bien el modelo explica los datos (0 a 1, donde 1 es ajuste perfecto).
- **R² ajustado (Adj. R-squared)** → similar al R², pero penaliza por incluir demasiadas variables irrelevantes.
- **F-statistic y su valor p** → prueba conjunta para ver si el modelo en general es significativo.

### 🔍 Otros datos
- **Número de observaciones (No. Observations)** → cantidad de datos usados en la regresión.
- **Grados de libertad (Df Residuals, Df Model)** → información sobre restricciones en el modelo.
- **Durbin-Watson** → prueba para detectar autocorrelación en los residuos.
- **Jarque-Bera (JB), Skew, Kurtosis** → pruebas sobre la normalidad de los residuos.

---
👉 En resumen, `lr.summary()` es una forma compacta y completa de evaluar la calidad del modelo, la importancia de cada predictor y la validez estadística de los resultados.

## DETECTING COLLINEARITY WITH EIGENVECTORS

- Análisis matemático de la matriz de covarianza (o correlación) de tus variables predictoras para descubrir combinaciones lineales casi redundantes.

```python
eigenvalues, eigenvectors = np.linalg.eig(df.corr())

pd.options.display.float_format = '{:,.4f}'.format
pd.Series(eigenvalues).sort_values()
```

- Podemos ver la varianza de los datos; cuanto mayor, es el numero mas se estiran los datos; cuanto menor, direcciones donde casi no hay variacion. Cuanto mas cercanos son los valores a 0, señalan colinealidad, lo que indica que son variables que aportan informacion casi identica. (Menores a 0.2)

- Numero de condicion -> varianza mayor / varianza menor Un numero de condicion muy alto indica que los coeficientes de regresion son muy inestables, pequeños cambios en los datos podrian generar grandes alteracione

```python
# Extrameos uno de los eigenvectores (RAD) de la matriz y luego has ordenado sus componentes por magnitud. Cada componente de ese vector es un peso que indica cuánto “entra”
    #cada variable en esa combinación lineal
np.abs(pd.Series(eigenvectors[:,8])).sort_values(ascending = False)
```

## 📌 Conclusión práctica sobre la multicolinealidad

- **TAX y RAD están altamente correlacionadas** → al tener coeficientes grandes y similares, prácticamente aportan la **misma información** al modelo.  
- Esto es un claro ejemplo de **multicolinealidad**, donde dos o más variables explicativas están fuertemente relacionadas, dificultando separar su efecto individual.  

### 🔎 Consecuencias de la multicolinealidad
- Los **valores p** de los coeficientes se vuelven poco confiables.  
- Los **coeficientes pueden variar mucho** si se agrega o elimina una variable correlacionada.  
- TAX y RAD "compiten" por explicar la misma parte de la varianza, dejando a otras variables (INDUS, NOX, LSTAT, etc.) con menos peso.  

### ✅ Qué hacer en la práctica
1. **Eliminar una de las variables** (TAX o RAD) si representan lo mismo en términos conceptuales.  
2. **Combinar las variables correlacionadas** en un índice o componente.  
3. Usar técnicas robustas frente a multicolinealidad, como **Ridge** o **Lasso Regression**.  

---
👉 En resumen: **el modelo actual está sobre-explicando con dos variables que son casi idénticas. Corregir esto mejora la estabilidad y la interpretación de los resultados.**


```python
# Los tres primeros resultados tienen una carga alta comparados con el resto. Esto quiere decir que son los facotres que nos estan 
    # causando multicolinealidad
print(df.columns[2], df.columns[8], df.columns[9])
```

## 🔑 Resumen

- Incluir **TAX, RAD e INDUS** directamente en el modelo **no mejora la capacidad predictiva global**,  
  pero sí provoca que sus **coeficientes individuales sean poco fiables** debido a la multicolinealidad.  

- Para resolverlo, es clave **romper la redundancia** entre estas variables. Algunas estrategias son:  
  - **Regularización** (Ridge o Lasso) → penaliza coeficientes grandes y estabiliza el modelo.  
  - **PCA (Análisis de Componentes Principales)** → combina variables correlacionadas en factores independientes.  
  - **Agregación** → crear un índice o métrica compuesta que resuma TAX, RAD e INDUS.  

👉 De esta forma, se conserva la **información valiosa de las tres variables** sin perder la  
**estabilidad ni la interpretabilidad** de los coeficientes del modelo.

## STANDARDISE VARIABLE TO IDENTIFY KEY FEATURES

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x, y)
```

```python
# Construimos un DataFrame que asocia cada coeficiente con el nombre de la columna correspondiente
# Usamos list() -> para envolver al zip(), que es un iterador que solo se puede recorrer una vez,de forma que con el list podemos pasar una estructura completa
# zip() -> empareja el coeficiente con el nombre de la variable
# set_index() -> seleccionamos que columna se tomara como indice de la fila
result = pd.DataFrame(list(zip(model.coef_, df.columns)), columns = ['coeficient', 'name']).set_index('name')
# Ordenamos los coeficientes
np.abs(result).sort_values(by = 'coeficient', ascending = False)
```

```python
# Lo mismo de antes pero normalizando
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline

# Creamos un objeto scaler que sabe como centrar y escalar datos
scaler = StandardScaler()
# Aplicamos una tuberia que primero aplica scaler y luego sobre el resultado entrena al modelo
standard_coef_linear_reg = make_pipeline(scaler, model)

# Entrenamos el pipeline completo
standard_coef_linear_reg.fit(x, y)
# Extraemos los coeficientes del segundo paso del pipeline (steps[1]) que es nuestro LinearRegresion, el cual contiene una tupla por eso se añade otro [1], para
   # quedarnos con el modelo ya entrenado con datos estandarizados
result = pd.DataFrame(list(zip(standard_coef_linear_reg.steps[1][1].coef_, df.columns)), columns = ['coeficient', 'name']).set_index('name')
np.abs(result).sort_values(by = 'coeficient', ascending = False)
```

## USE r^2 TO IDENTIFY KEY FEATURES

- Forma muy sencilla de medir “qué tanto aporta cada variable”

```python
from sklearn.metrics import r2_score

linear_reg = smf.ols(formula = 'y ~ CRIM + ZN + CHAS + NOX + RM + AGE + DIS + RAD + TAX + PTRATIO + B + LSTAT', data = df)
benchmark = linear_reg.fit()
r2_score(y, benchmark.predict(df))
```

```python
# Quitamos LSTAT
linear_reg = smf.ols(formula = 'y ~ CRIM + ZN + CHAS + NOX + RM + AGE + DIS + RAD + TAX + PTRATIO + B', data = df)
benchmark = linear_reg.fit()
r2_score(y, benchmark.predict(df))
```

- ¿Qué nos dice eso? Que LSTAT aportaba una buena cantidad de “poder explicativo”. Al quitarla, el modelo ya no predice tan bien.

```python
# Quitamos la edad
linear_reg = smf.ols(formula = 'y ~ CRIM + ZN + CHAS + NOX + RM + DIS + RAD + TAX + PTRATIO + B + LSTAT', data = df)
benchmark = linear_reg.fit()
r2_score(y, benchmark.predict(df))
```

- ¿Qué significa? Que AGE casi no aportaba información extra una vez que ya tenía las otras variables. Su ausencia no empeora el ajuste.
